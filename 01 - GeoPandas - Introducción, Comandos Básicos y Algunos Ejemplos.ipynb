```python
localidades_shp["area"] = localidades_shp.area
localidades_shp["area"]
```

```python
localidades_shp["perimetro"] = localidades_shp.boundary
localidades_shp
```

```python
localidades_shp["centroide"] = localidades_shp.centroid
localidades_shp
```

```python
punto_inicial = localidades_shp["centroide"].iloc[0]
```

```python
punto_inicial
```

```python
localidades_shp["distancia"] = localidades_shp["centroide"].distance(punto_inicial)
localidades_shp["distancia"]
```

```python
localidades_shp["distancia"].mean()
```

```python
localidades_shp.plot("area", legend = True)
```

```python
localidades_shp = localidades_shp.set_geometry("centroide")
```

```python
localidades_shp.plot("area", legend = True)
```

```python
axis = localidades_shp["geometry"].plot()
localidades_shp["centroide"].plot(ax = axis, color = "black")
```

```python
localidades_shp.plot(column = "CODIGO_LOC", legend = True, legend_kwds = {"loc": "upper left"}, figsize = (20, 12))
```

```python
localidades_shp["envolvente"] = localidades_shp.convex_hull
```

```python
axis = localidades_shp["envolvente"].plot(alpha = 0.5)
localidades_shp["perimetro"].plot(ax = axis, color = "white", linewidth = 0.5)
```

```python
localidades_shp["buffer"] = localidades_shp.buffer(10000)
```

```python
localidades_shp["buffer_centroide"] = localidades_shp["centroide"].buffer(10000)
```

```python
axis = localidades_shp["buffer"].plot(alpha = 0.5)
localidades_shp["buffer_centroide"].plot(ax = axis, color = "red", alpha = 0.5)
localidades_shp["perimetro"].plot(ax = axis, color = "white", linewidth = 0.5)
```
